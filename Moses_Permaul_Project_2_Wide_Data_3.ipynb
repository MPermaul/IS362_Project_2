This data set contains some statistics and details on daily trending YouTube videos for the US. The data keeps track of various statistics such as likes and dislikes, comments, and descriptions. The data is daily details, so there is a large amount of data.

Some analysis that can be performed is to find out what videos are trending the longest and how the rate of likes and or dislikes increases the longer a video is on the trending list. 

After cleaning this data, we will find the following:
1. What videos are trending the longest
2. The rate of likes and/or dislikes the longer a video is trending

### Python Code for Imports and Reading the Data
To begin, we will import the standard libraries needed, read in the data, and display the first 5 rows.

````python
# standard imports for numpy and pandas
import numpy as np
import pandas as pd

# read wide data to DataFrame
raw_data = pd.read_csv('data/wide_data_3.csv')

# make a copy of DataFrame to preserve origanal import
wide_data = raw_data.copy()

# view first 5 rows to understand the data
wide_data.head()
````

### Cleaning the Data
In order to answer our questions, we won't need the majority of the columns currently in the DataFrame. We will start by removing the ones we don't need.

````python
# wide_data[wide_data.title == 'WE WANT TO TALK ABOUT OUR MARRIAGE']

# get column list
wide_data.columns

# delete columns not neede
wide_data.drop(columns=['category_id', 'publish_time', 'tags', 'views', 'comment_count', 'thumbnail_link', 
                        'comments_disabled', 'ratings_disabled','video_error_or_removed', 'description'], inplace=True)

# view first 5 rows
wide_data.head()
````

We will now look at the data types within the "wide_data" DataFrame in order to make sure that we have the right data types.

````python
# view data types for the DataFrame
wide_data.dtypes
````

The first thing that we notice is that the column "**treding_date**" is a string and not date.To clean this, we will **convert** it to datetime.

````python
# convert the 'trending_date' column to datetime format, errors param set to 'coerce'
wide_data['trending_date'] = pd.to_datetime(wide_data['trending_date'], format='%y.%d.%m', errors='coerce')

# view first 5 rows
wide_data.head()
````

Looking at the data types now, we should now have clean data to work with.

````python
# view data types for the DataFrame
wide_data.dtypes
````

### Analyzing the Data

1. What videos are trending the longest?

    To find this, we will create a new dataframe that stores each title and a count of the title's video id. The count of the video id will let us know how any days the title has been trending for.

````python
# create DataFrame that has the video titles and count video_id
df_count = wide_data.groupby(['title']).agg(['count'])['video_id']

# use nlargest on count to view top 10
df_count.nlargest(10, 'count')
````

Looking at the data, we can see that the title "**WE MADE OUR MOM CRY...HER DREAM CAME TRUE!**" trended the longest. It was on the list for 30 different days

2. The rate of likes and/or dislikes the longer a video is trending.

Since the ***data contains the over 6455 unique titles***:

```python
wide_data['title'].nunique()
```
We are going to focus on only the top title.

````python
# create DataFrame with data from most trending video
df_top = wide_data.loc[wide_data['title'] == 'WE MADE OUR MOM CRY...HER DREAM CAME TRUE!']

# update DataFrame with sort values on trending date
df_top = df_top.sort_values('trending_date')
````

````python
# view full dataframe
df_top
````

````python
# add column for percent change in likes
df_top['pct_like_change'] = df_top['likes'].pct_change()

# add column for percent change in dislikes
df_top['pct_dislikes_change'] = df_top['dislikes'].pct_change()
````

Using the percent change in likes, we will view first 5 days which had the biggest percent increase.

````python
# sort dataframe by pct_like_change
df_top = df_top.sort_values('pct_like_change', ascending=False)

# view first 5 rows
df_top[['trending_date', 'likes', 'dislikes', 'pct_like_change']].head()
````

Using the percent change in dislikes we will view first 5 days which had the biggest percent increase.

````python
# sort dataframe by pct_like_change
df_top = df_top.sort_values('pct_dislikes_change', ascending=False)

# view first 5 rows
df_top[['trending_date', 'likes', 'dislikes', 'pct_dislikes_change']].head()
````

Looking at the results from both percentages, we can see that biggest changes happened at the beginning of the trending period. The rate of change for dislikes were higher at some points, the number of likes were higher overall.
